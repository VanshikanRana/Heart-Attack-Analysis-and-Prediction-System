# Predictive Modelling

```python
# Convert categorical variables to numeric variable
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cat_var=['Gender','Has Diabetes', 'Smoking Status', 'Chest Pain Type', 'Treatment']
for var in cat_var:
    le.fit(heart_data[var])
    heart_data[var]=le.transform(heart_data[var])
```

```python
# Import the Libraries 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Feature Engineering
feature=['Gender', 'Age', 'Blood Pressure (mmHg)', 'Cholesterol (mg/dL)','Has Diabetes', 'Smoking Status', 'Chest Pain Type']
x=heart_data[feature]
y=heart_data.Treatment
```

```python
# Data Splitting
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# Create & Train the Model
model=LogisticRegression(max_iter=1000,multi_class='ovr')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
# Precdicting the Treatment
sample=[]
sex=int(input("Enter Gender of Patient (0: FEMALE & 1: MALE): "))
sample.append(sex)
age=int(input("Enter Age of Patient : "))
sample.append(age)
BP=int(input("Enter the Blood Pressure (mmHg) of Patient :"))
sample.append(BP)
Ch=int(input("Enter the Cholesterol (mg/dL) of Patient :"))
sample.append(Ch)
DB=int(input("Does Patient has Diabetes ( 0: NO & 1: YES) : "))
sample.append(DB)
smoke=int(input("Does Patient smoke\n0: CURRENT \n1: FORMER \n2: NEVER : "))
sample.append(smoke)
cpt=int(input("Enter The Type of Chest Pain of Patient: \n 0:'Asymptomatic'\n 1:'Atypical Angina'\n 2:'Non-anginal Pain'\n 3:'Typical Angina' :"))
sample.append(cpt)
sample=np.array([sample])
prediction=(model.predict(sample))
print('\n')
if (prediction==0):
    print('Type of Treatment is : Angioplasty')
elif (prediction==1):
    print('Type of Treatment is : Coronary Artery Bypass Graft (CABG)')
elif (prediction==2):
    print('Type of Treatment is : Lifestyle Changes')
else :
    print('Type of Treatment is : Medication')
print("It is just a Prediction :)\n")
print(f'Accuray: {accuracy_score(y_test,y_pred)}\n')
print(classification_report(y_test,y_pred))
```
